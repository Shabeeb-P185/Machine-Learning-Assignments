```python
import pandas as pd

# Read the CSV file
df = pd.read_csv("food_db.csv")

# Print number of rows and columns
print("Shape of dataframe:", df.shape)

# Show the dataframe
print(df)
```

```python
# Replace 5% and 10% discounts with 13%
new_df = df.replace({'discount': {'5%': '13%', '10%': '13%'}})

# Show the updated dataframe
print(new_df)
```

```python
# Replace categorical rating values with numeric values
rating_map = {
    'Excellent': 4,
    'very Good': 3,
    'Good': 2,
    'Average': 1
}

new_df = df.replace({'rating': rating_map})

# Show the dataframe
print(new_df)
```
